```python
def mat(imagepath):
    imagepath = imagepath.split('.jpg')[0]
    imagepath = imagepath + '.mat'
    data = sio.loadmat(imagepath)
    landmarks = data['pt3d_68']
    labels = data['Pose_Para'][0][:3]
    pitch, yaw, roll = labels
    return landmarks, pitch, yaw, roll
```

```python
x_yp=[]
features1=[]
labesl1=[]
imname=[]
def get_points(imagepath):
    mm = sio.loadmat(imagepath)
    pos = mm['Pose_Para'][0][:3]*180/np.pi #Yaw pitch roll
    points = mm['pt3d_68']
    bins = np.array(range(-99, 102, 3))
    labels =np.digitize(pos, bins) - 1 
    x_yp.append(np.concatenate((points[0],points[1])))
    features1.append(pos)
    labesl1.append(labels)
```

```python
def draw_axis_1(img, yaw, pitch, roll, tdx=None, tdy=None, size = 100):

    pitch = pitch * np.pi / 180
    yaw = -(yaw * np.pi / 180)
    roll = roll * np.pi / 180

    if tdx != None and tdy != None:
        tdx = tdx
        tdy = tdy
    else:
        height, width = img.shape[:2]
        tdx = width / 2
        tdy = height / 2

    # X-Axis pointing to right. drawn in red
    x1 = size * (cos(yaw) * cos(roll)) + tdx
    y1 = size * (cos(pitch) * sin(roll) + cos(roll) * sin(pitch) * sin(yaw)) + tdy

    # Y-Axis | drawn in green
    #        v
    x2 = size * (-cos(yaw) * sin(roll)) + tdx
    y2 = size * (cos(pitch) * cos(roll) - sin(pitch) * sin(yaw) * sin(roll)) + tdy

    # Z-Axis (out of the screen) drawn in blue
    x3 = size * (sin(yaw)) + tdx
    y3 = size * (-cos(yaw) * sin(pitch)) + tdy

    cv2.line(img, (int(tdx), int(tdy)), (int(x1),int(y1)),(0,0,255),3)
    cv2.line(img, (int(tdx), int(tdy)), (int(x2),int(y2)),(0,255,0),3)
    cv2.line(img, (int(tdx), int(tdy)), (int(x3),int(y3)),(255,0,0),2)

    return img
```

```python
def draw_axis(img, yaw, pitch, roll, tdx=None, tdy=None, size = 60):
    if tdx != None and tdy != None:
        tdx = tdx
        tdy = tdy
    else:
        height, width = img.shape[:2]
        tdx = width / 2
        tdy = height / 2

    # X-Axis pointing to right. drawn in red
    x1 = size * (cos(yaw) * cos(roll)) + tdx
    y1 = size * (cos(pitch) * sin(roll) + cos(roll) * sin(pitch) * sin(yaw)) + tdy

    # Y-Axis | drawn in green
    #        v
    x2 = size * (-cos(yaw) * sin(roll)) + tdx
    y2 = size * (cos(pitch) * cos(roll) - sin(pitch) * sin(yaw) * sin(roll)) + tdy

    # Z-Axis (out of the screen) drawn in blue
    x3 = size * (sin(yaw)) + tdx
    y3 = size * (-cos(yaw) * sin(pitch)) + tdy

    cv2.line(img, (int(tdx), int(tdy)), (int(x1),int(y1)),(0,0,255),3)
    cv2.line(img, (int(tdx), int(tdy)), (int(x2),int(y2)),(0,255,0),3)
    cv2.line(img, (int(tdx), int(tdy)), (int(x3),int(y3)),(255,0,0),2)

    return img
```

```python
def show_image(imagepath):
    capture = cv2.VideoCapture(imagepath)
    ret, frame = capture.read()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    landmarks, pitch, yaw, roll = mat(imagepath)

    x,y = landmarks[0,32], landmarks[1,32]
    plt.imshow(draw_axis(frame, -yaw, pitch, roll, x, y))
    plt.show()

    capture.release()
    cv2.destroyAllWindows()
```
